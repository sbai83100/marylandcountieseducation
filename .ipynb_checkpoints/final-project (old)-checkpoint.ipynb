```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import LabelEncoder
```

### Enrollment

```python
enrollment = pd.read_csv("Maryland_Public_School_Enrollment_Trends__Pre-Kindergarten_to_Grade_12__2009-2019.csv")
county_names = enrollment.columns[3:len(enrollment.columns)-1]
enrollment_melted = pd.melt(enrollment, id_vars=['Year', "Grades"], value_vars=county_names)
enrollment_melted = enrollment_melted.rename(columns={'variable': "County", 'value': "Enrollment"})
enrollment_2018 = enrollment_melted[(enrollment_melted["Grades"] == "9-12") & (enrollment_melted["Year"] == 2018)]
enrollment_2018 = enrollment_2018.sort_values(by='County')

enrollment_2018.drop(columns=["Year", "Grades"], inplace=True)
enrollment_2018
```

### Spending

```python
spending = pd.read_csv("County_Board_of_Education_-_Spending_Disclosures.csv")
spending = spending.replace({'CALVERT COUNTY PUBLIC SCHOOLS': "Calvert County Public Schools", 
                  'DORCHESTER COUNTY PUBLIC SCHOOLS': "Dorchester County Public Schools",
                  'Howard  County Public Schools': 'Howard County Public Schools', 
                  'Montgomery County Public School': 'Montgomery County Public Schools',
                  'Oakland, Garrett County Public Schools': 'Garrett County Public Schools'})
agencies_old = spending["Agency Name"].unique()
agencies_old.sort()
agencies_new = ['Allegany County', 'Anne Arundel County', 'Baltimore City',
       'Baltimore County', 'Calvert County', 'Caroline County',
       'Carroll County', 'Cecil County', 'Charles County',
       'Dorchester County', 'Frederick County', 'Garrett County',
       'Harford County', 'Howard County', 'Kent County',
       'Montgomery County', "Prince George's County",
       "Queen Anne's County", 'Somerset County', "St. Mary's County",
       'Talbot County', 'Washington County', 'Wicomico County',
       'Worcester County']

res = dict(zip(agencies_old, agencies_new))
spending = spending.replace(res)
spending = spending.rename(columns={"Agency Name": "County"})


spending_totals = spending.groupby(["Fiscal Year", "County"]).sum()
spending_totals = spending_totals.reset_index()
spending_totals = spending_totals[~(spending_totals["Amount"] == 0)]
spending_totals[spending_totals["Fiscal Year"] == 2018]
```

### Population

```python
population = pd.read_csv("Maryland_Resident_Population_Per_Square_Mile__2010-2019.csv")
population = population.drop(columns=["Date created", "MARYLAND"])
county_names = np.array(population.columns[1:])

population_melted = pd.melt(population, id_vars=['Year'], value_vars=county_names)
population_melted = population_melted.rename(columns={'variable': "County", 'value': "Population Per Square Mile"})
population_melted[population_melted["Year"] == 2018]
```

### Poverty

```python
poverty = pd.read_csv("Poverty_Rate_With_Margin_Of_Error__2010-2018.csv")
poverty = poverty.drop(columns=["Date Created"])
moe = poverty[poverty["Estimate"] == "MOE"]
poverty = poverty[~poverty["Estimate"].str.contains("MOE")]
poverty = poverty.drop(columns=["Estimate"])
county_names = poverty.columns[2:]
poverty_melted = pd.melt(poverty, id_vars=['Year'], value_vars=county_names)
poverty_melted = poverty_melted.rename(columns={'variable': "County", 'value': "Poverty Rate"})
poverty_melted#[poverty_melted["County"] == "Baltimore City"]
```
